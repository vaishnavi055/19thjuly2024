```python
import pandas as pd
import json


file_path = '/content/DataEngineeringQ2 (1).json'
with open(file_path) as file:
    data = json.load(file)
patientDetails = pd.json_normalize(data)


print("Columns in the DataFrame:", patientDetails.columns)


def percentage_missing(df, column):
    total_count = len(df)
    missing_count = df[column].apply(lambda x: isinstance(x, str) and x.strip() == '').sum()
    return round((missing_count / total_count) * 100, 2)


columns_to_check = ['patientDetails.firstName', 'patientDetails.lastName', 'patientDetails.DOB']


percentages = []
for column in columns_to_check:
    if column in patientDetails.columns:
        perc_missing = percentage_missing(patientDetails, column)
        percentages.append(perc_missing)
    else:
        print(f"Column '{column}' not found in the DataFrame")
        percentages.append(0.00)
print(', '.join(f'{percentage:.2f}' for percentage in percentages))
```

```python
import pandas as pd
import json
from datetime import datetime

# Load the JSON file into a DataFrame
file_path = '/content/DataEngineeringQ2 (1).json'  # Replace with your actual file path
with open(file_path) as file:
    data = json.load(file)
df = pd.json_normalize(data)

# Function to calculate age from birthdate
def calculate_age(birthdate):
    if pd.isna(birthdate) or birthdate.strip() == '':
        return None
    birthdate = datetime.strptime(birthdate, '%Y-%m-%dT%H:%M:%S.%fZ')
    today = datetime.today()
    return today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))

# Add the age column
df['patientDetails.age'] = df['patientDetails.birthDate'].apply(calculate_age)

# Function to categorize age into age groups
def categorize_age(age):
    if age is None:
        return 'Unknown'
    elif age <= 12:
        return 'Child'
    elif age <= 19:
        return 'Teen'
    elif age <= 59:
        return 'Adult'
    else:
        return 'Senior'

# Add the ageGroup column
df['patientDetails.ageGroup'] = df['patientDetails.age'].apply(categorize_age)

# Count the number of Adults
adult_count = (df['patientDetails.ageGroup'] == 'Adult').sum()

# Print the result
print(adult_count)
```

```python
import pandas as pd

# Load the JSON dataset into a DataFrame
df = pd.read_json('/content/DataEngineeringQ2 (1).json')

# Define a function to check if a phone number is valid
def is_valid_phone_number(phone_number):
    # Remove spaces and leading/trailing whitespace
    phone_number = phone_number.strip()

    # Check if the number starts with '+91' or '91'
    if phone_number.startswith('+91'):
        phone_number = phone_number[3:]  # Remove '+91'
    elif phone_number.startswith('91'):
        phone_number = phone_number[2:]  # Remove '91'

    # Check if the remaining number is exactly 10 digits
    if len(phone_number) == 10 and phone_number.isdigit():
        number = int(phone_number)
        return 6000000000 <= number <= 9999999999

    return False

# Apply the validation function to create the 'isValidMobile' column
df['isValidMobile'] = df['phoneNumber'].apply(is_valid_phone_number)

# Print the number of valid phone numbers
valid_count = df['isValidMobile'].sum()
print(f'Number of valid phone numbers: {valid_count}')

# Optional: Print the DataFrame to verify the new column
print("\nDataFrame with 'isValidMobile' column:")
print(df.head())
```
